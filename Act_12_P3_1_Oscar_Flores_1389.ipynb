## Formulas
* $M = A > B$

```python
# Verdadero (si)
# Calcular cual de los dos numeros es el mayor usando if else
# variables : A, B, M
# Formula M = A > B
A = int(input("Ingrese el primer numero: "))
B = int(input("Ingrese el segundo numero: "))
if A > B:
    M = A
else:
    M = B
print("El numero mayor es: ", M)
print("Fin del programa")
print("Programa realizado por Oscar Flores NL = 19")
```

```python
# Falso (no)
# Calcular cual de los dos numeros es el mayor usando if else
# variables : A, B, M
# Formula M = A > B
A = int(input("Ingrese el primer numero: "))
B = int(input("Ingrese el segundo numero: "))
if A > B:
    M = A
else:
    M = B
    print("El numero mayor es: ", M)
print("Fin del programa")
print("Programa realizado por Oscar Flores NL = 19")
```
